```python
from __future__ import print_function, division

%matplotlib inline

from com.sbk.signal.square.square_signal import SquareSignal


from com.sbk.converter.converter import signal_to_wave
from com.sbk.converter.converter import wave_to_spectrum
from IPython.display import display
```

```python


square_wave_frequency=50
signal = SquareSignal(square_wave_frequency)
duration = signal.period*6
sbksegment = signal_to_wave(signal, duration, frame_rate=10000)
sbksegment.plot()
```

```python
wave = signal_to_wave(signal, duration=0.5, frame_rate=10000)
wave.apodize()
wave.make_audio()
```

```python


spectrum = wave_to_spectrum(wave)
spectrum.plot(high=2000)
```
