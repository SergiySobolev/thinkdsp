```python
from __future__ import print_function, division

%matplotlib inline

from com.sbk.signal.triangle.triangle_signal import TriangleSignal
from com.sbk.signal.triangle.sbk_triangle_signal import SbkTriangleSignal


from com.sbk.converter.converter import signal_to_wave
from com.sbk.converter.converter import wave_to_spectrum
from IPython.display import display
```

```python

triangle_wave_frequency=200
sbksignal = SbkTriangleSignal(triangle_wave_frequency)
duration = sbksignal.period*6
sbksegment = signal_to_wave(sbksignal, duration, frame_rate=10000)
sbksegment.plot()
```

```python
signal = TriangleSignal(triangle_wave_frequency)
duration = sbksignal.period*6
segment = signal_to_wave(signal, duration, frame_rate=10000)
segment.plot()
```

```python
wave = signal_to_wave(signal, duration=0.5, frame_rate=10000)
wave.apodize()
wave.make_audio()
```

```python


spectrum = wave_to_spectrum(wave)
spectrum.plot(high=2000)
```
