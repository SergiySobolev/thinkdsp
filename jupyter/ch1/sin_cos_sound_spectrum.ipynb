```python
from __future__ import print_function, division

%matplotlib inline

from com.sbk.dsp import sbkdsp

from IPython.display import display
```

```python
f1 = 100
a1 = 30
o1 = 0

f2 = 200
a2 = 80
o2 = 0

cos = sbkdsp.cos_signal(f1,a1,o1)
sin = sbkdsp.sin_signal(f2,a2,o2)
```

```python
cos.plot()
```

```python
sin.plot()
```

```python
s = cos + sin
s.plot()
```

```python
wave = s.make_wave(duration=3, start=0, frame_rate=11025)
```

```python
subwave = wave.segment(0,0.2)
subwave.plot()
```

```python
spec = wave.make_spectrum()
spec.plot()
```

```python
spectrogram = wave.make_spectrogram(1)
spectrogram.plot()
```
