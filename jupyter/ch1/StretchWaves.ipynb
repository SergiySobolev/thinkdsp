```python
from __future__ import print_function, division

%matplotlib inline

from com.sbk.dsp import sbkdsp
from com.sbk.wave.sbkwave import SbkWave
import com.sbk.func.trigonometry.trigonometry as trigonometry

from IPython.display import display
```

```python
test_sinus_freq = 880
test_sinus_amp = 1
test_sinus_offset = 0
test_frame_rate = 24000
test_duration = 3
        
```

```python
sinus_signal = trigonometry.sin_signal(freq=test_sinus_freq,
                                            amp=test_sinus_amp,
                                            offset=test_sinus_offset)
            
```

```python
sinus_signal.plot(frame_rate=test_frame_rate, periods = 10)
```

```python
from com.sbk.converter.converter import signal_to_wave

base_sinus_wave = SbkWave(wave=signal_to_wave(sinus_signal, duration=test_duration, frame_rate=test_frame_rate))
base_segment = base_sinus_wave.segment(start=0, duration=sinus_signal.period*10)
base_segment.plot()                       
```

```python
stretched_signal = base_segment.stretch(2)
stretched_signal.plot()
```
