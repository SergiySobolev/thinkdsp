```python
%pylab inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
Fs = 150.0;  # sampling rate
Ts = 1.0/Fs; # sampling interval
t = np.arange(0,1,Ts) # time vector
fq = 10;   # frequency of the signal
y = np.sin(2*np.pi*fq*t)
```

```python
n = len(y) # length of the signal
k = np.arange(n)
T = n/Fs
frq = k/T # two sides frequency range

n2 = n // 2
frq = frq[0:n2] # one side frequency range
Y = np.fft.fft(y)/n # fft computing and normalization
Y = Y[0:n2]
```

```python
plt.plot(t,y)
plt.title('Original Signal = sin(2*PI*10*t)', color='blue')
plt.xlabel('t')
plt.ylabel('sin(2*PI*10*t)')
plt.show()

plt.plot(frq,abs(Y))
plt.xlabel('Freq (Hz)')
plt.ylabel('|Y(freq)|')
plt.show()
```

```python
Y_shifted = np.fft.fftshift(np.fft.fft(y)/n)
Y_shifted_half = Y_shifted[0:n2]
plt.plot(frq, abs(Y_shifted_half))
plt.xlabel('Freq (Hz) shifted to center')
plt.ylabel('|Y(freq)|')
plt.show()
```
